## Viz - Bokeh

```python
df_AP['Date'] =  pd.to_datetime(df_AP['Date'], format='%d-%B-%Y')
df_AP_sorted = df_AP.sort_values('Date')
df_AP_sorted
```

```python
from math import pi
# from bokeh.sampledata.glucose import data

day = df_AP_sorted.loc['244':'0']

p = figure(x_axis_type="datetime", title="Traded Days", plot_height=400, plot_width=1000)
p.xaxis.formatter.days = '%d/%m/%Y'
p.xaxis.major_label_orientation = pi/3

p.line(day.Date, day['Close Price'])

cr = p.circle(day.Date, day['Close Price'], size=10,
              fill_color="grey", hover_fill_color="firebrick",
              fill_alpha=0.05, hover_alpha=0.3,
              line_color=None, hover_line_color="white")
p.add_tools(HoverTool(tooltips=None, renderers=[cr], mode='hline'))


show(p)
```
